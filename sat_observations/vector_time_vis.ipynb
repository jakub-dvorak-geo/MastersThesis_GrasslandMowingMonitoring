```python
def parse_date(datestr):
    print(datestr.split('_'))

def datestr_to_doy(year, datestr):
    day, month = datestr.split('_')
    return date(year, int(month), int(day)).timetuple().tm_yday
```

```python
obs_doy = set(datestr_to_doy(year, datestr) for datestr in datestrs)
```
